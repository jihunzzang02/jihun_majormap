```python
import pickle
from gensim.models import Word2Vec
from tqdm import tqdm
from loader import FileLoader
import operator
import pandas as pd
import re
from konlpy.tag import Okt
import os
okt=Okt()

from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize


#각 학과를 doc2vec을 돌려서 학과당 256차원의 벡터를 부여함
def doc2vec_making_256vector():
    
    with open('C:/Users/jihun/OneDrive/바탕 화면/GitHub/nexus/_artifacts_/notebooks/jihun/us_univ_wordbag_not_filtered',"rb") as fr:
        us_univ_wordbag_not_filtered = pickle.load(fr)
    with open('C:/Users/jihun/OneDrive/바탕 화면/GitHub/nexus/_artifacts_/notebooks/jihun/us_univ_name',"rb") as fr:
        depts = pickle.load(fr)
    
        
    tagged_text = [TaggedDocument(words=word_tokenize(d.lower()), tags=[str(i)]) for i,d in enumerate(us_univ_wordbag_not_filtered)]
    
    max_epochs = 100
    vec_size = 256
    alpha = 0.025

    model = Doc2Vec(size=vec_size,
                    alpha=alpha, 
                    min_alpha=0.00025,
                    min_count=1,
                    dm =1)

    model.build_vocab(tagged_text)

    for epoch in range(max_epochs):
        print('iteration {0}'.format(epoch))
        model.train(tagged_text,
                    total_examples=model.corpus_count,
                    epochs=model.iter)
        # decrease the learning rate
        model.alpha -= 0.0002
        # fix the learning rate, no decay
        model.min_alpha = model.alpha

    model.save("d2v.model")
    print("Model Saved")
    
    
    
    model= Doc2Vec.load("d2v.model")
    
    
    vector_arr={}
    for i in range(len(us_univ_wordbag_not_filtered)):    
        vector_arr[depts[i]['university']+' Univ / ' +depts[i]['department']]=model.infer_vector(word_tokenize(us_univ_wordbag_not_filtered[i].lower()))

    with open('C:/Users/jihun/OneDrive/바탕 화면/GitHub/nexus/_artifacts_/notebooks/jihun/us_univ_vector_256',"wb") as fw:
        pickle.dump(vector_arr,fw)
    
    
    
    
    
    
    

```
